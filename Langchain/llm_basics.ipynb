# **To combine the Chain and  to set a sequence for that we use SimpleSequentialChain**

```python
prompt_template_name = PromptTemplate(
    input_variables =['cuisine'],
    template = "I want to open a restaurant for {cuisine} food. Suggest a fency name for this."
)

name_chain =LLMChain(llm=llm, prompt=prompt_template_name)
```

```python
prompt_template_items = PromptTemplate(
    input_variables = ['restaurant_name'],
    template="""Suggest some menu items for {restaurant_name}"""
)

food_items_chain = LLMChain(llm=llm, prompt=prompt_template_items)
```

```python
from langchain.chains import SimpleSequentialChain
chain = SimpleSequentialChain(chains = [name_chain,food_items_chain])
```

```python

content = chain.run("indian")
print(content)
```

# SequentialChain

```python

prompt_template_name = PromptTemplate(
    input_variables =['cuisine'],
    template = "I want to open a restaurant for {cuisine} food. Suggest a fency name for this."
)

name_chain =LLMChain(llm=llm, prompt=prompt_template_name, output_key="restaurant_name")
```

```python


prompt_template_items = PromptTemplate(
    input_variables = ['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}."
)

food_items_chain =LLMChain(llm=llm, prompt=prompt_template_items, output_key="menu_items")
```

```python
from langchain.chains import SequentialChain
```

```python

chain = SequentialChain(
    chains = [name_chain, food_items_chain],
    input_variables = ['cuisine'],
    output_variables = ['restaurant_name', "menu_items"]
)
```

```python
print(chain({"cuisine": "indian"}))
```

# Agent

```python
from langchain.agents import initialize_agent, Tool
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
```

```python
def search_tool(query):
    # Imagine this function performs a search operation
    return f"Results for: {query}"

tools = [Tool(name="search_tool", func=search_tool, description="Search for information")]

# Step 3: Initialize the agent
agent = initialize_agent(tools=tools, llm=llm, agent_type="zero-shot-react-description",handle_parsing_errors=True)
```

```python
user_input = "What are the latest advancements in iran and israel war?"
response = agent.run(user_input)
print(response)
```

```python
from langchain.agents import AgentType, initialize_agent, load_tools
from langchain.llms import OpenAI
```

```python
agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose = True,
    handle_parsing_errors=True
)
```

```python
agent.run("What was the GDP of us in 2022?")
```

# Memory

```python
from langchain.prompts import PromptTemplate

prompt_template_name = PromptTemplate(
    input_variables =['cuisine'],
    template = "I want to open a restaurant for {cuisine} food. Suggest a fency name for this."
)
```

```python
from langchain.chains import LLMChain

chain = LLMChain(llm=llm,prompt=prompt_template_name)
name = chain.run("indian")
print(name)
```

```python
name = chain.run("usa")
print(name)
```

```python
name = chain.run("australia")
print(name)
```

```python
chain.memory
```

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template_name, memory=memory)
name = chain.run("India")
print(name)
```

```python
name = chain.run("usa")
print(name)
```

```python
name = chain.run("australia")
print(name)
```

```python
print(chain.memory.buffer)
```

```python
from langchain.chains import ConversationChain

convo = ConversationChain(llm=llm)
print(convo.prompt.template)
```

```python
convo.run("what is  the older name of india")
```

```python
convo.run("How much is 5+5?")
```

```python
print(convo.memory.buffer)
```

```python
from langchain.memory import ConversationBufferWindowMemory
memory = ConversationBufferWindowMemory(k=3)


convo = ConversationChain(
    llm=llm,
    memory=memory
)
convo.run("Who won the first cricket world cup?")
```

```python
convo.run("How much is 5+5?")
```

```python
convo.run("Who was the captain of the winning team?")
```
