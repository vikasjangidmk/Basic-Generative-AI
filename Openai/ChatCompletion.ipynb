```python
prompt = "who are you?"

response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ]
)
response
```

```python
prompt1 = "Hello How are you?"
prompt2 = "I am 25 years old & I am a software engineer."
prompt3 = "Tell me about me"

response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt1},
        {"role": "user", "content": prompt2},
        {"role": "user", "content": prompt3},
    ]
)

response['choices'][0]['message']['content']
```

```python
prompt = "What is Python?"

response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ],
    max_tokens = 25
)
response['choices'][0]['message']['content']
```

```python
prompt = "What is Python?"

response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ],
    max_tokens = 5,
    temperature = 0.5,
    top_p = 0.95,
    n=1,
    stop=None,
    frequency_penalty=0,
    presence_penalty=0
)
response['choices'][0]['message']['content']
```

```python
prompt = "Give me hindi translation of this sentence: This movie is Amazing."

response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ],
    temperature = 0.5,
)
response['choices'][0]['message']['content']
```

## **Image Generation**

```python
response = openai.Image.create(
    prompt = "A painting of a bird",
    n=1,
    size = "256x256"
)

# The URL of the image
image_url = response['data'][0]['url']

# Output the image URL
print(image_url)
```

## Embedding

```python
# Define the text you want to embed
text  = "OpenAI develops advanced AI models"

response = openai.Embedding.create(
    model = "text-embedding-ada-002", #One of the most powerful and efficient embedding models
    input = text
)

# Extract the embedding from the response
embedding = response['data'][0]['embedding']

# Output the embedding
print(f"Embedding vector: {embedding}")
```
